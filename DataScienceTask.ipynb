### Feature Extractor Implementation
Since we deal with different approaches to feature extraction, we provide an abstract base class for all concrete implementations.

It ensures flexibility in case we need to implement a new feature extraction method.

```python
class BaseFeatureExtractor(ABC):
    """ Abstract Base Class for feature extraction """
    
    def __init__(self, sample_rate, chunk_duration):
        """
        Initializes the BaseFeatureExtractor
        
        Args:
            sample_rate (int): Sampling rate of the audio data 
            chunk_duration (float): Duration in sec of each audio chunk to process
        """
        self.sample_rate = sample_rate
        self.chunk_duration = chunk_duration
        
    @abstractmethod
    def _extract_feature(self, audio):
        """
        Abstract method for feature extraction.
        Concrete classes must implement this method to extract features.

        Args:
            audio (numpy.ndarray): 1D array of raw audio samples
        Returns:
            numpy.ndarray: Extracted features from the audio
        """
        pass

    def extract_feature_for_segment(self, audio_path):
        """
        Extracts features for a specific segment of an audio file
        
        Args:
            audio_path (str): Path to the audio file

        Returns:
            numpy.ndarray: Extracted features (or None if loading fails)
        """
        # Load the audio segment (by default, from the middle)
        audio_data = self._load_segment(audio_path, self.chunk_duration)

        if (audio_data is None):
            return None

        # Extract features from the loaded audio data
        features = self._extract_feature(audio_data)

        return features
        
    def extract_feature_by_chunk(self, audio_path):
        """
        Extracts features from an audio file by splitting it into chunks and processing each chunk
        
        Args:
            audio_path (str): Path to the audio file.
        
        Returns:
            numpy.ndarray: Array of extracted features for each chunk of the audio
        """
        # Load the entire audio data from audio_path
        audio = self._load(audio_path) 

        # Convert chunk duration to number of samples based on the sampling rate
        chunk_size = self.chunk_duration * self.sample_rate

        # Calculate the number of chunks
        num_chunks = int(np.ceil(len(audio) / chunk_size))

        feature_list = []

        for i in range(num_chunks):
            start = i * chunk_size   # Calculate the start sample of the chunk
            end = start + chunk_size # Calculate the end sample of the chunk
    
            # Extract chunk, if it's shorter than required duration it's discarded
            # Another approach could be zero-padding if necessary
            chunk = audio[start:end]
            if len(chunk) < chunk_size:
                    break

            # Extract features for the current chunk
            features = self._extract_feature(chunk)
            feature_list.append(features)

        # Convert the list of feature arrays to a NumPy array
        # The final shape of the feature array is (num_chunks, num_features, num_observations)
        feature_list = np.array(feature_list)
        return feature_list

    def _load(self, audio_path):
        """
        Loads the audio from a given path and applies peak normalization
    
        Args:
            audio_path (str): Path to the audio file
    
        Returns:
            numpy.ndarray: Normalized audio data as a 1D NumPy array 
        """
        # load 1-channel audio with predefined sampling rate
        audio = librosa.load(audio_path, sr=self.sample_rate, mono=True)[0]

        # Apply peak normalization to the audio
        audio = self._peak_normalize(audio)

        return audio
        
    def _load_segment(self, audio_path, duration, from_middle = True):
        """
        Loads a specific segment of an audio file.
    
        The segment can either be from the beginning or the middle of the audio,
        based on the `from_middle` flag. If the audio file is shorter than the 
        required duration, `None` is returned.
    
        Args:
            audio_path (str): Path to the audio file
            duration (float): Duration (in seconds) of the audio segment
            from_middle (bool): Whether to extract the segment from the middle of the audio
    
        Returns:
            numpy.ndarray: Normalized audio data as a 1D NumPy array 
        """
        
        # Load the entire audio
        audio = self._load(audio_path) 

        # Calculate the number of required samples based on the duration and sample rate
        required_samples = int(duration * self.sample_rate) 
        total_samples = len(audio)

        # If shorter than required duration, return None
        if total_samples <= required_samples:
            return None

        # If not from_middle, return the audio samples from the beginning
        if not from_middle:
            return audio[:required_samples]

        # Otherwise, extract the segment from the middle of the audio
        # This can help avoid silence or unwanted noise at the beginning of the audio
        start_sample = (total_samples - required_samples) // 2 
        return audio[start_sample:start_sample + required_samples]

    def _peak_normalize(self, audio):
        """
        Normalizes the audio by scaling its peak amplitude to 1
    
        Args:
            audio (numpy.ndarray): The audio data to normalize
    
        Returns:
            numpy.ndarray: The normalized audio data
        """
        peak = np.max(np.abs(audio))  # Find the absolute max value
        if peak == 0:
            return audio  # Avoid division by zero
        return audio / peak  # Scale (the peak is now 1)

class MFCCExtractor(BaseFeatureExtractor):
    """ Concrete class for extracting MFCC (Mel-Frequency Cepstral Coefficients) """
    
    def __init__(self, sample_rate, number_mfcc, chunk_duration):
        """
        Initializes the MFCCExtractor

        Args:
            sample_rate (int): Sampling rate of the audio data
            number_mfcc (int): Number of MFCC coefficients to extract
            chunk_duration (float): Duration in sec of each audio chunk to process
        """
        super().__init__(sample_rate, chunk_duration)
        self.number_mfcc = number_mfcc
        
    def _extract_feature(self, audio_data):
        """
        Extracts MFCC features from the provided audio data.

        Args:
            audio_data (numpy.ndarray): 1D array of raw audio samples

        Returns:
            numpy.ndarray: Extracted MFCC features, a 2D array (number_mfcc x time_frames)
        """
        mfccs = librosa.feature.mfcc(y=audio_data, sr=self.sample_rate, n_mfcc=self.number_mfcc)
        return mfccs

class Wav2VecExtractor(BaseFeatureExtractor):
    """ Concrete class for feature extraction using the wav2vec model """
    
    def __init__(self, path_to_model, sample_rate, chunk_duration):
        """
        Initializes the Wav2VecExtractor and loads the model in evaluation mode.
    
        Args:
            path_to_model (str): Path to the pretrained wav2vec model checkpoint
            sample_rate (int): Sampling rate of the input audio
            chunk_duration (float): Duration in sec of each audio chunk to process
        """
        super().__init__(sample_rate, chunk_duration)
        
        # Load pretrained wav2vec model
        model, cfg, task = fairseq.checkpoint_utils.load_model_ensemble_and_task([path_to_model])
        model = model[0]
        model.eval()
        self.model = model
        
    def _extract_feature(self, audio_data):
        """
        Extracts wav2vec features from raw audio data
        
        Args:
            audio_data (numpy.ndarray): 1D array of raw audio samples
    
        Returns:
            numpy.ndarray: Extracted features from the wav2vec model
        """
        # Convert to Torch tensor
        audio_tensor = torch.from_numpy(audio_data)

        # Adds one dimension to make it suitable for wav2Vec model
        audio_tensor = audio_tensor.unsqueeze(0) 

        # Extract features using wav2vec model
        wav2Vec_features = self.model.feature_extractor(audio_tensor)

        # Remove one dimension and convert tensor back to NumPy array
        wav2Vec_features_np = wav2Vec_features.squeeze(0).detach().numpy()

        # Converting between Torch tensors and NumPy arrays is not very efficient, 
        # but it's done for compatibility with the project's base architecture.
        return wav2Vec_features_np
```

### Dataset builder implementation

We need to build two types of datasets: one for gender classification and another for speaker classification.

The original dataset is fairly balanced, so each speaker is more or less equally represented.

For **gender classification**, we split the recordings by both gender and speaker, ensuring that the training dataset contains an equal number of male and female speakers. Additionally, speakers in the training set do not appear in the test set.

For the **speaker classification** task, all speakers must be present in both, the training and test sets. To achieve this, we split the data by speaker and by recordings, ensuring that each speaker appears in both sets, but with different recordings for training and testing.

```python
class BaseDatasetBuilder(ABC):
    """ Abstract class for dataset creation """
    
    def __init__(self, dataset_path, speakers_path):
        """
        Initializes the dataset builder with paths for the dataset and speakers 

        Args:
            dataset_path (str): Path to the dataset directory
            speaker_path (str): Path to the speakers' information
        """
        self.dataset_path, self.speaker_path = dataset_path, speaker_path
        
        self.X_train = self.y_train = self.X_test = self.y_test = None
        
        self._male_ids = []
        self._female_ids = []

        # Flag indicating whether the dataset is ready
        # To avoid rebuilding it from scratch
        self._is_ready = False
        
        self._parse_dataset()

    @property
    def ready(self):
        return self._is_ready
        
    @property
    def male_ids(self):
        return self._male_ids

    @property
    def female_ids(self):
        return self._female_ids

    @abstractmethod
    def build_dataset(self, feature_extractor):
        """
        Abstract method that must be implemented by subclasses to build the dataset

        Args:
            feature_extractor: Feature extractor object used to extract features from audio files
        """
        pass
        
    def get_dataset(self):
        """
        Returns the dataset if it was built before

        Raises:
            ValueError: If the dataset isn't ready yet (i.e., build_dataset hasn't been called).

        Returns:
            tuple: A tuple containing (X_train, y_train, X_test, y_test) 
        """
        if self._is_ready:
            return (self.X_train, self.y_train, self.X_test, self.y_test)
        else:
            raise ValueError("The dataset isn't ready yet, use <build_dataset> method instead")

    def get_speaker_rec_paths(self, speaker_id):
        """
        Retrieves all audio file paths for a given speaker

        Args:
            speaker_id (str): The ID of the speaker

        Returns:
            list: List of file paths to the audio recordings of the speaker
        """

        # Directory path for the given speaker ID
        id_folder = os.path.join(self.dataset_path, speaker_id)

        # Find all .flac files under the speaker's folder
        speaker_rec_paths = glob.glob(id_folder + "/**/*.flac", recursive=True)
        
        return speaker_rec_paths
        
    def _parse_dataset(self):
        """
        Parses the dataset to find male and female speaker IDs
    
        Raises:
            ValueError: If the dataset or speaker file is not found
        """
        if not os.path.exists(self.dataset_path) or not os.path.exists(self.speaker_path):
            raise ValueError('Required files not found. Check the dataset folder.')
        
        # Parse .txt file with speakers' info
        # Read speaker data (ID, Gender): n rows x 2 columns, where 1st column is ID, 2nd column is a gender(M or F)
        df = pd.read_csv(self.speaker_path, sep="|", comment=";", skipinitialspace=True, usecols=[0, 1], header=None)
        
        # Trim whitespaces
        df = df.map(lambda x: x.strip() if type(x)==str else x)
    
        all_speakers_ids = os.listdir(self.dataset_path)
        
        # Get ids of male speakers 
        # note: 7976 - wrong Male
        self._male_ids = [speaker_id for speaker_id in all_speakers_ids if self._parse_gender(df, speaker_id) == "M"]

        # Get ids of female speakers
        self._female_ids = [speaker_id for speaker_id in all_speakers_ids if self._parse_gender(df, speaker_id) == "F"]

    def _parse_gender(self, df_speakers, speaker_id):
        """
        Identifies the gender of a speaker based on their ID from the speaker dataframe
    
        Args:
            df_speakers (pandas.DataFrame): DataFrame containing speaker IDs and genders
            speaker_id (int): Speaker's ID
    
        Returns:
            str: The gender of the speaker, 'M' for male or 'F' for female.
            None: If no gender information is found for the speaker.
        """
        if not speaker_id.isdigit():
            return None

        speaker_id = int(speaker_id)
        result = df_speakers.loc[df_speakers[0] == speaker_id, 1]
        return result.values[0] if not result.empty else None
        
class SpeakerDatasetBuilder(BaseDatasetBuilder):
    """ Builds dataset for speaker classification task """

    def build_dataset(self, extractor, split = 0.8):
        """
        Builds the datasets by splitting each speaker's recordings into train/test sets
        
        Args:
            extractor: Feature extractor
            split (float, optional): Proportion of data used for training
        """
        
        X_train, y_train, X_test, y_test = [], [], [], []

        # Combine male and female speaker IDs
        all_speakers = self._male_ids + self._female_ids
        speaker_index = 0
        
        for speaker_id in all_speakers:
            print("Collecting data for speaker: ", speaker_id)
            # Get speaker's recording paths
            speaker_rec_paths = self.get_speaker_rec_paths(speaker_id)

            # Compute the number of the speaker's recordings for training according to the split
            speaker_train_rec_num = int(split * len(speaker_rec_paths))

            # Randomly choose recordings for training
            speaker_train_rec_paths = random.sample(speaker_rec_paths, speaker_train_rec_num)

            # Remaining for testing
            speaker_test_rec_paths = [rec_path for rec_path in speaker_rec_paths if rec_path not in speaker_train_rec_paths]

            # Extract features for the training set
            X_train_cur = self._extract_features_from_paths(speaker_train_rec_paths, extractor)
            X_train.append(X_train_cur)
            y_train.extend([speaker_index]*len(X_train_cur))

            # Extract features for the test set
            X_test_cur = self._extract_features_from_paths(speaker_test_rec_paths, extractor)
            X_test.append(X_test_cur)
            y_test.extend([speaker_index]*len(X_test_cur))
            
            speaker_index+=1

        # Convert lists to arrays and print dataset shapes
        self.X_train, self.y_train = np.vstack(X_train), np.array(y_train)
        self.X_test, self.y_test = np.vstack(X_test), np.array(y_test)
        
        # Print shapes of train and test datasets
        print(f"Train dataset shape: {self.X_train.shape}, {self.y_train.shape}")
        print(f"Test dataset shape: {self.X_test.shape}, {self.y_test.shape}")
        
        # Mark the dataset as ready
        self._is_ready = True
        
        return (self.X_train, self.y_train, self.X_test, self.y_test)

    def _extract_features_from_paths(self, rec_paths, extractor):
        """
        Extracts features from audio recordings given their paths

        Args:
            rec_paths (list): A list of paths to the audio recordings
            extractor (BaseFeatureExtractor): A feature extractor object used to extract features from audios

        Returns:
            numpy.ndarray: Extracted features of an audio recording
        """
        X = []

        # Iterate through each recording path to extract features
        for path in rec_paths:
            # Extract features for the current recording path
            features = extractor.extract_feature_by_chunk(path)
            
            if features is not None and len(features) > 0:
                X.append(features)

        # Convert the list of extracted features to numpy array
        # In our case, it will be 3D array (num_chunks, num_features, num_observations)
        X = np.vstack(X)
        return X
        
class GenderDatasetBuilder(BaseDatasetBuilder):
    """ Builds dataset for gender classification task """
        
    def build_dataset(self, extractor, split = 0.8):
        """
        Builds the datasets by splitting male and female speakers 
        for train and test sets according to the split.
        
        Args:
            extractor: Feature extractor
            split (float, optional): Proportion of data used for training
        """

        # Calculate number of speakers for training and testing
        train_male_speakers_num = int(split * len(self._male_ids))
        train_female_speakers_num = int(split * len(self._female_ids))

        # randomly choose male speakers for test and train according to the split
        train_male_speakers = random.sample(self._male_ids, train_male_speakers_num)
        test_male_speakers = [male_id for male_id in self._male_ids if male_id not in train_male_speakers]

        # randomly choose female speakers for test and train according to the split
        train_female_speakers = random.sample(self._female_ids, train_female_speakers_num)
        test_female_speakers = [female_id for female_id in self._female_ids if female_id not in train_female_speakers]

        # Extract features for male and female speakers 
        X_train_female = self._extract_features_for_ids(train_female_speakers, extractor)
        y_train_female = np.ones((len(X_train_female),), dtype=int) 
        X_test_female = self._extract_features_for_ids(test_female_speakers, extractor)
        y_test_female = np.ones((len(X_test_female),), dtype=int) 
         
        X_train_male = self._extract_features_for_ids(train_male_speakers, extractor)
        y_train_male = np.zeros((len(X_train_male),), dtype=int) 
        X_test_male = self._extract_features_for_ids(test_male_speakers, extractor)
        y_test_male = np.zeros((len(X_test_male),), dtype=int) 

        # Concatenate male and female data for training and testing
        X_train = np.concatenate((X_train_female, X_train_male), axis=0) 
        y_train = np.concatenate((y_train_female, y_train_male), axis=0)  
        
        X_test = np.concatenate((X_test_female, X_test_male), axis=0)
        y_test = np.concatenate((y_test_female, y_test_male), axis=0)

        # Shuffle training data 
        X_train, y_train = shuffle(X_train, y_train, random_state=42)
        
        # Print dataset shapes for debugging
        print(f"Female data: Train {X_train_female.shape}, Test {X_test_female.shape}")
        print(f"Male data: Train {X_train_male.shape}, Test {X_test_male.shape}")
        print(f"Train dataset shape: {X_train.shape}, {y_train.shape}")
        print(f"Test dataset shape: {X_test.shape}, {y_test.shape}")

        self.X_train, self.y_train, self.X_test, self.y_test = X_train, y_train, X_test, y_test

        # Mark the dataset as ready
        self._is_ready = True
        
        return (X_train, y_train, X_test, y_test)

    def get_male_features(self, extractor, only_segment = False):
        """
        Extract features for male speakers

        Args:
            extractor (BaseFeatureExtractor): Feature extractor
            only_segment (bool, optional): Whether to extract features from the entire audio or a segment
            
        Returns:
            np.array: The extracted features for male speakers
        """
        X_male = self._extract_features_for_ids(self._male_ids, extractor, only_segment)
        return X_male

    def get_female_features(self, extractor, only_segment = False):
        """
        Extract features for female speakers

        Args:
            extractor (BaseFeatureExtractor): Feature extractor
            only_segment (bool, optional): Whether to extract features from the entire audio or a segment
            
        Returns:
            np.array: The extracted features for female speakers
        """
        X_female = self._extract_features_for_ids(self._female_ids, extractor, only_segment)
        return X_female

    def _extract_features_for_ids(self, speaker_ids, extractor, only_segment = False):
        """
        Extract features for a list of speaker IDs

        Args:
            speaker_ids (list): List of speaker IDs to process
            extractor (BaseFeatureExtractor): The feature extractor
            only_segment (bool, optional): Whether to extract features from the entire audio or a segment

        Returns:
            np.array: The extracted features for the provided speakers
        """
        X = []
        for speaker_id in speaker_ids:
            # Get the paths of recordings for the current speaker
            speaker_rec_paths = self.get_speaker_rec_paths(speaker_id)

            for path in speaker_rec_paths:
                # Extract features based on the `only_segment` flag
                if only_segment:
                    features = extractor.extract_feature_for_segment(path)
                    if features is not None and len(features) > 0:
                        X.append(features)
                else:
                    features = extractor.extract_feature_by_chunk(path)
                    if features is not None and len(features) > 0:
                        X.extend(features)
                        
        # Convert the list of extracted features to a numpy array          
        X = np.array(X)
        return X
```

### Features Distribution

MFCC (Mel-Frequency Cepstral Coefficients) is a commonly used feature in speech recognition, representing the vocal tract as a small number of coefficients. It is particularly useful for capturing the timbral characteristics of speech. In this case, we aim to extract these features and check any potential correlation with gender.

To achieve this, we will extract a number of MFCC features for both male and female speakers and analyze how they are distributed. The most important features are the first 13 MFCCs, so we will focus on plotting their distributions.

We will extract MFCC features in chunks, considering 2-second audio fragments. For each group (males and females), the resulting feature arrays will have the following shape: (number_of_chunks, number_of_mfccs, number_of_observations). The extraction process takes some time, but ultimately, we have 4,217 MFCC features for males and 4,175 MFCC features for females. This indicates that, with respect to recording durations, the dataset is fairly balanced.

```python
SR = 16000
CHUNK_DURATION = 2
NUM_MFCC = 13

extractor_obj = MFCCExtractor(SR, NUM_MFCC, CHUNK_DURATION)

print("Computing MFCCs for male recordings...")
male_mfccs = dataset_obj.get_male_features(extractor_obj)

print("Computing MFCCs for female recordings...")
female_mfccs = dataset_obj.get_female_features(extractor_obj)

print("Male MFCCs shape: ", male_mfccs.shape)
print("Female MFCCs shape: ", female_mfccs.shape)
```

### Visualising distribution of MFCCs

We have computed 13 MFCCs for 63 time frames (number of observations). Since we are interested in the MFCCs themselves, we need to reshape these feature arrays to a shape of (num_chunks * num_observations, 13).

From the distribution plots, we observe that the MFCCs show considerable overlap between male and female speakers. However, some MFCCs (specifically, 2, 11, and 13) shows noticeable shifts in mean values between males and females. This could indicate that males and females tend to have higher energy in certain frequency ranges. Additionally, the distributions appear to have slightly different variances, which could suggest patterns in male and female speech.

These observations suggest that MFCCs can be useful for classification tasks. However, for future analyses, we will extract 39 MFCC components instead of the initial 13.

```python
# Reshape to (num_chunks*num_obs, num_mfccs)
if male_mfccs.ndim == 3:
    male_mfccs = male_mfccs.transpose(0, 2, 1).reshape(-1, NUM_MFCC)
print("Male MFCCs shape after reshaping: ", male_mfccs.shape)

if female_mfccs.ndim == 3:
    female_mfccs = female_mfccs.transpose(0, 2, 1).reshape(-1, NUM_MFCC)
print("Female MFCCs shape after reshaping: ", female_mfccs.shape)

print("\nPlotting MFCCs distribution by gender...")

# Setting up the figure and the subplots
fig, axes = plt.subplots(nrows=5, ncols=3, figsize=(15, 12))
axes = axes.flatten()

# Plot each MFCC distribution
for i in range(NUM_MFCC):
    sns.kdeplot(male_mfccs[:, i], label="Male", color="blue", fill=True, ax=axes[i])
    sns.kdeplot(female_mfccs[:, i], label="Female", color="red", fill=True, ax=axes[i])
    
    axes[i].set_title(f"MFCC {i+1} Distribution")
    axes[i].set_xlabel("MFCC Value")
    axes[i].set_ylabel("Density")
    axes[i].legend()

# Hide unused subplots
axes[-1].axis("off")
axes[-2].axis("off")

plt.tight_layout()
plt.show()
```

### Visualisation with boxplots

Let's now visualize the MFCC coefficients using boxplots. The boxplots display the distribution of the first 5 MFCCs for male and female speakers. ## Training classes

This module contains the training logic for both classical ML methods and DL models. It includes the classifier classes as well as the PytorchTrainer, which implements the PyTorch training procedure.

PytorchTrainer handles the training and evaluation process, incorporating an early stopping mechanism and saving the best model to the local drive.

```python
class PytorchTrainer():
    """ PyTorch training process implementation """
    
    def __init__(self, model, model_name, output_dir):
        """
        Initialize the trainer with the model, model name, and output directory for checkpoints

        Args:
            model (torch.nn.Module): The PyTorch model to train
            model_name (str): The name of the model, used to name the checkpoint
            output_dir (str): Directory where the model checkpoints will be saved
        """
        self.model = model
        self.model_name = model_name
        self.train_dataset = self.test_dataset = self.val_dataset = None
        self.optimizer = None
        self.criterion = None
        self.hyperparameters = None
        self.path_to_checkpoint = os.path.join(output_dir, f'{self.model_name}')
    
    def set_hyperparameters(self, optimizer, criterion, lr, epochs, batch_size, early_stop_patience):
        """
        Set the hyperparameters for training the model.

        Args:
            optimizer (torch.optim.Optimizer): Optimizer class to use
            criterion (torch.nn.Module): Loss function to use
            lr (float): Learning rate for the optimizer
            epochs (int): Number of epochs to train the model
            batch_size (int): Size of the batches used in training
            early_stop_patience (int): Number of epochs to wait before stopping training early if no improvement
        """
        self.hyperparameters = {
            'lr': lr,
            'epochs': epochs,
            'batch_size': batch_size,
            'early_stop_patience': early_stop_patience
        }
        self.optimizer = optimizer(self.model.parameters(), lr=lr)
        self.criterion = criterion
    
    def train(self, X_train, y_train, X_val, y_val):
        """
        Train the model using the given training and validation data

        Args:
            X_train (np.array): Training dataset
            y_train (np.array): Training dataset labels
            X_val (np.array): Validation dataset
            y_val (np.array): Validation dataset labels

        Raises:
            ValueError: If the datasets are not properly initialized

        Functionality:
            - Initializes datasets and data loaders
            - Iterates over multiple epochs to train the model
            - Tracks loss and accuracy for both training and validation datasets
            - Implements early stopping to prevent overfitting
            - Saves the best model based on validation loss
            - Restores the best-performing model after training
        """

        # Create datasets
        self.train_dataset = SpeechDataset(X_train, y_train)
        self.val_dataset = SpeechDataset(X_val, y_val)

        train_loss_per_epoch = []
        train_acc_per_epoch = []
        val_loss_per_epoch = []
        val_acc_per_epoch = []

       # Ensure the datasets are properly initialized
        if self.train_dataset is None or self.val_dataset is None:
            raise ValueError('Data is not prepared for training')

        # Retrieve hyperparameters
        epochs = self.hyperparameters['epochs']
        batch_size = self.hyperparameters['batch_size']
        early_stop_patience = self.hyperparameters['early_stop_patience']

        optimizer = self.optimizer
        criterion = self.criterion

        # Create DataLoader instances for batch processing
        train_dataloader = DataLoader(self.train_dataset, batch_size=batch_size, shuffle=True, drop_last=True)
        val_dataloader = DataLoader(self.val_dataset, batch_size=batch_size, shuffle=False, drop_last=True)

        # Initialize early stopping variables
        best_val_loss = np.inf
        early_stop_counter = 0
        epsilon = .001

        # Training loop over epochs
        for epoch in range(epochs):
            total_acc_train = 0  # accumulate training accuracy
            total_loss_train = 0 # accumulate training loss

            # Training phase
            for batch in tqdm(train_dataloader, desc=f'Epoch {epoch + 1}/{epochs}', disable = True):
                inputs, labels = batch

                # Forward pass: get model predictions
                outputs = self.model(inputs)
                                      
                # Compute the loss
                loss = criterion(outputs, labels.squeeze())

                # Accumulate total loss
                total_loss_train += loss.item()

                # Compute accuracy
                acc = (outputs.argmax(dim=1) == labels.squeeze()).sum().item()
                total_acc_train += acc

                # Backpropagation and optimization step
                optimizer.zero_grad()
                loss.backward()
                optimizer.step()
    
            total_acc_val = 0  # Accumulate validation accuracy
            total_loss_val = 0 # Accumulate validation loss

            # Validation phase (disable gradient computation)
            with torch.no_grad():
                for batch in val_dataloader:
                    inputs, labels = batch

                    # Forward pass
                    outputs = self.model(inputs)

                    # Compute loss
                    loss = criterion(outputs, labels.squeeze())

                    # Accumulate validation loss
                    total_loss_val += loss.item()

                    # Compute accuracy
                    acc = (outputs.argmax(dim=1) == labels.squeeze()).sum().item()
                    
                    total_acc_val += acc

            
            train_len = len(self.train_dataset)
            val_len = len(self.val_dataset)

            # Print epoch summary
            print(
               f'Epochs: {epoch + 1}/{epochs} | Train Loss: {total_loss_train / train_len: .5f} | Train Accuracy: {total_acc_train / train_len: .5f} | Val Loss: {total_loss_val / val_len: .5f} | Val Accuracy: {total_acc_val / val_len: .5f}')

            # Early Stopping mechanism implementation
            # Check if validation loss has improved
            if total_loss_val < best_val_loss - epsilon:
                # Save model checkpoint as the current best model
                self._save_model_checkpoint(epoch + 1)
                print(f'Saved model checkpoint at epoch {epoch + 1} to {self.path_to_checkpoint}')

                # Update the best validation loss
                best_val_loss = total_loss_val
                
                # Reset the early stopping counter
                early_stop_counter = 0
            else:
                # If no improvement, increase the early stopping counter
                early_stop_counter += 1

            # If validation loss hasn't improved for a certain number of epochs, stop training
            if early_stop_counter >= early_stop_patience:
                print(f"Triggered early stopping at epoch {epoch}")
                break
            print()
    
        # Load the best model from the checkpoint
        print('Restoring the best model: ')
        checkpoint = torch.load(self.path_to_checkpoint + ".pth")
        self.model.load_state_dict(checkpoint['model_state_dict'])
        restored_epoch = checkpoint['epoch']
        
        print(f'The best model state was restored from {restored_epoch} epoch')

    def evaluate(self, X_test, y_test):
        """
        Evaluates the model's performance on the test dataset
        
        Args:
            X_test (np.array): The input features for the test set.
            y_test (np.array): The true labels for the test set.

        Returns:
            dict: A dictionary containing the evaluation metrics
        """

        # Create a SpeechDataset instance for the test data
        self.test_dataset = SpeechDataset(X_test, y_test)
        
        batch_size = self.hyperparameters['batch_size']

        # Create a DataLoader for batching the test dataset
        test_dataloader = DataLoader(self.test_dataset, batch_size = batch_size)

        # Initialize variables to accumulate the total loss, accuracy, and to store labels and predictions
        total_loss_test = 0
        total_acc_test = 0
        all_labels = []
        all_preds = []

        # Disable gradient computation during evaluation
        with torch.no_grad():
            # Iterate through the test dataset in batches
            for batch in tqdm(test_dataloader, desc='Testing model', disable=True):
                inputs, labels = batch

                # Forward pass: get model predictions for the batch
                outputs = self.model(inputs)

                # Compute the loss between the model's predictions and true labels
                loss = self.criterion(outputs, labels.squeeze())

                # Accumulate the total loss
                total_loss_test += loss.item()

                # Compute accuracy
                acc = (outputs.argmax(dim=1) == labels.squeeze()).sum().item()
                total_acc_test += acc

                # Store the true labels and predicted labels for confusion matrix calculation
                all_labels.extend(labels.squeeze().numpy())
                all_preds.extend(outputs.argmax(dim=1).numpy())

        # Compute the confusion matrix
        conf_matrix = confusion_matrix(all_labels, all_preds)

        # Compute precision, recall, and F1 score for the model
        precision, recall, f1, _ = precision_recall_fscore_support(all_labels, all_preds, average='weighted')

        report = {
            'test_loss': total_loss_test / len(self.test_dataset),
            'test_accuracy': total_acc_test / len(self.test_dataset),
            'cm': conf_matrix,
            'precision': precision,
            'recall': recall,
            'f1_score': f1
        }
        return report

    def _save_model_checkpoint(self, epoch):
        """
        Save the current state of the model to a checkpoint for future use.
        It saves a checkpoint file (.pth) containing the model state dictionary and the current epoch.

        Args:
            epoch (int): The current epoch number. This will be saved in the checkpoint file.
        """
        
        path_to_model_checkpoint = self.path_to_checkpoint + ".pth"

        # Get the directory name from the path
        dir_name = os.path.dirname(path_to_model_checkpoint)

        os.makedirs(dir_name, exist_ok=True)

        torch.save({
            'epoch': epoch,
            'model_state_dict': self.model.state_dict()
        }, path_to_model_checkpoint)

```

### Dataset for DL models

Prepares data suitable for the PyTorch format.

```python
class SpeechDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)  # Convert X to tensor
        self.y = torch.tensor(y, dtype=torch.long)  # Convert y to tensor (for classification)

    def __len__(self):
        return len(self.X)  # Number of samples

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]  # Return a single sample (X, y)
```

### Classifiers Implementation

Here, we implement the base class for all classifiers to maintain a uniform pipeline. Additionally, we provide implementations for two machine learning classifiers: Gaussian Nave Bayes and Random Forest. For deep learning models, we use a Multilayer Perceptron (MLP) with three dense layers and a Convolutional Neural Network (CNN) with two convolutional layers.

```python
class BaseClassifier(ABC):
    """ Abstract class for classification models. """

    @abstractmethod
    def init_model(self, input_dim, num_classes):
        pass
        
    @abstractmethod
    def train(self, X_train, y_train):
        pass
    
    @abstractmethod
    def evaluate(self, X_test, y_test):
        pass

    def show_cm(self, cm):
        # Confusion matrix visualization
        plt.figure(figsize=(6, 5))
        sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
        
        plt.xlabel("Predicted Label")
        plt.ylabel("True Label")
        plt.title("Confusion Matrix")
        plt.show()
        
class MLClassifier(BaseClassifier):
    """ Machine Learning Classifier """
    
    def __init__(self, classifier_type):
        """
        Initializes the ML classifier.

        Args:
            classifier_type (str): Specifies the type of classifier to use 
        """
        self._classifier_type = classifier_type
        self._classifier = None
        self._classifier_name = ""
        self._report = None

    def init_model(self, input_shape, num_classes):
        """ Initializes the classifier based on the specified type """
        
        if (self._classifier_type == "random_forest"):
            self._classifier = RandomForestClassifier()
            self._classifier_name = "Random Forest"
        elif (self._classifier_type == "naive_bayes"):
            self._classifier = GaussianNB()
            self._classifier_name = "Gaussian Nave Bayes"
        else:
            raise ValueError('The classifier is not defined')
        
    def train(self, X_train, y_train):
        """
        Trains the classifier using the provided training data.

        Args:
            X_train (np.array): Training data
            y_train (np.array): Labels for training data
        """
         # Flatten the feature matrix if it has more than two dimensions
        if X_train.ndim  > 2:
            X_train = X_train.reshape(X_train.shape[0], -1)

        self._classifier.fit(X_train, y_train)
    
    def evaluate(self, X_test, y_test):
        """
        Evaluates the classifier's performance on the test data

        Args:
            X_test (np.array): Test data
            y_test (np.array): True labels for the test data
        """
        
        # Flatten the feature matrix if it has more than two dimensions
        if X_test.ndim  > 2:
            X_test = X_test.reshape(X_test.shape[0], -1)

        # Predict labels using the trained classifier
        y_pred = self._classifier.predict(X_test)

        # Compute metrics
        precision, recall, fscore, _ = precision_recall_fscore_support(y_test, y_pred, average='weighted')
        accuracy = accuracy_score(y_test, y_pred)

        # Store evaluation results
        self._report = {
            'accuracy': accuracy,
            'precision': precision,
            'recall': recall,
            'f1_score': fscore
        }

        # Compute confusion matrix
        self._cm = confusion_matrix(y_test, y_pred)
        
    def visualise(self):
        print(self._classifier_name + " performance: ")
        
        # Convert metrics to a list for tabulate
        table = [self._report.keys(), self._report.values()]
        
        # Print table
        print(tabulate(table, headers="firstrow", tablefmt="fancy_grid"))
        
        self.show_cm(self._cm)
        
class DLClassifier(BaseClassifier):
    """ Deep Learning Classifier base class """

    def __init__(self, output_folder):
        """
        Initializes the deep learning classifier

        Args:
            output_folder (str): Directory where model checkpoints and outputs are saved
        """
        self.model = None
        self._model_name = ""
        self._report = None
        self._trainer = None
        self.output_folder = output_folder
        
    def _init_trainer_parameters(self):
        """
        Initializes hyperparameters for the model training process
        """
        # Initialize hyperparameters
        epochs = 300
        batch_size = 32
        lr = 1e-4
        optimizer = AdamW
        early_stop_patience = 20

        # Define loss function
        loss = nn.CrossEntropyLoss()

        # Set hyperparameters in the trainer instance
        self._trainer.set_hyperparameters(optimizer, loss, lr, epochs, batch_size, early_stop_patience)

    def train(self, X_train, y_train):
        """
        Trains the deep learning model.

        Args:
            X_train (np.array): Training data
            y_train (np.array): Corresponding labels for training data
        """
        # Splits the training data into training and validation sets (85% train, 15% validation).
        X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.15, random_state=42)

        # Train the model using the trainer instance
        self._trainer.train(X_train, y_train, X_val, y_val)
    
    def evaluate(self, X_test, y_test):
        """
        Evaluates the trained model on test data.

        Args:
            X_test (np.array): Test data
            y_test (np.array): True labels for the test data

        Returns:
            dict: Contains accuracy, precision, recall, F1-score, and confusion matrix.
        """
        self._report = self._trainer.evaluate(X_test, y_test)
        
    def visualise(self):
        """ Displays the evaluation results of the model, including a confusion matrix. """

        print(self._model_name + " performance: ")

        cm = self._report["cm"]
        
        table_report = self._report.pop("cm")
        
        # Convert metrics to a list for tabulate
        table = [self._report.keys(), self._report.values()]
        
        # Print table
        print(tabulate(table, headers="firstrow", tablefmt="fancy_grid"))

        self.show_cm(cm)

class MLPClassifier(DLClassifier):
    """ 
    Multi-Layer Perceptron (MLP) Classifier.
    This class implements an MLP model for deep learning classification.
    """
    def init_model(self, input_shape, num_classes):
        """
        Initializes the MLP model architecture.

        Args:
            input_shape (tuple): Shape of the input data
            num_classes (int): Number of output classes for classification

        Notes:
            - The input is flattened before being passed through fully connected layers.
            - Uses ReLU activation functions and dropout for regularization.
            - The final layer applies a Softmax activation for classification.
        """
        # Compute the input shape for the linear layer
        input_dim = input_shape[0] * input_shape[1]

        # Define the model architecture
        self.model = nn.Sequential(
            nn.Linear(input_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(64, num_classes),
            nn.Softmax(dim=1))

        self._model_name = "MLP model"

        # Initialize trainer with the model
        self._trainer = PytorchTrainer(self.model, "MLP_model", self.output_folder)

        # Set training hyperparameters
        self._init_trainer_parameters()

    def train(self, X_train, y_train):
        """
        Trains the MLP model.

        Args:
            X_train (np.array): Training data
            y_train (np.array): Corresponding labels for training data
        """
        # Reshape input if it has more than 2 dimensions.
        if X_train.ndim  > 2:
            X_train = X_train.reshape(X_train.shape[0], -1)

        # Split data into training and validation sets (85% train, 15% validation).
        X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.15, random_state=42)

        self._trainer.train(X_train, y_train, X_val, y_val)

    def evaluate(self, X_test, y_test):
        """
        Evaluates the trained MLP model on test data.

        Args:
            X_test (np.array): Test feature matrix.
            y_test (np.array): True labels for the test data.

        Returns:
            dict: Contains accuracy, precision, recall, F1-score, and confusion matrix.
        """
        # Flatten input data if it has more than 2 dimensions
        if X_test.ndim  > 2:
            X_test = X_test.reshape(X_test.shape[0], -1)

        # Evaluate the model
        self._report = self._trainer.evaluate(X_test, y_test)
        
class CNNClassifier(DLClassifier):
    """ Convolutional Neural Network (CNN) Classifier """
    
    def init_model(self, input_shape, num_classes):
        """
        Initializes the CNN model architecture.

        Args:
            input_shape (tuple): Shape of the input data
            num_classes (int): Number of output classes for classification
        """
        # Add a channel dimension (which is 1 in our case)
        input_dim = (1, input_shape[0], input_shape[1])

        # Define the CNN architecture
        self.model = nn.Sequential(
            nn.Conv2d(in_channels=input_dim[0], out_channels=32, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=3, stride=2, padding=1),
            nn.BatchNorm2d(32),
            nn.Conv2d(in_channels=32, out_channels=32, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=3, stride=2, padding=1),
            nn.Flatten(),
            nn.Linear(32 * math.ceil(input_dim[1] / 4) * math.ceil(input_dim[2] / 4), 64),  # Adjusted size after pooling
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(64, num_classes),
            nn.Softmax(dim=1))

        self._model_name = "CNN model"

        # Initialize trainer with the model
        self._trainer = PytorchTrainer(self.model, "CNN_model", self.output_folder)

        # Set training hyperparameters
        self._init_trainer_parameters()
    
    def train(self, X_train, y_train):
        """
        Trains the CNN model using the provided training data
    
        Args:
            X_train (np.array): Training data 
            y_train (np.array): Training labels
    
        Notes:
            - Splits the dataset into training and validation sets (85% train, 15% validation).
            - Expands dimensions to add a channel axis (required for CNN input).
            - Calls the trainer to execute the training process.
        """

        # Split the dataset into training and validation sets (85% train, 15% validation)
        X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.15, random_state=42)

        # Expand dimensions to add a channel axis (required for CNN input)
        X_train = np.expand_dims(X_train, axis=1)
        X_val = np.expand_dims(X_val, axis=1) 

        # Train the model using the trainer
        self._trainer.train(X_train, y_train, X_val, y_val)

    def evaluate(self, X_test, y_test):
        """
        Evaluates the trained CNN model on test data

        Args:
            X_test (np.array): Test data
            y_test (np.array): True labels for the test data

        Returns:
            dict: Contains accuracy, precision, recall, F1-score, and confusion matrix.
        """
        # Expand dimensions to match CNN input format
        X_test = np.expand_dims(X_test, axis=1)

        # Evaluate the model using the trainer and store the report
        self._report = self._trainer.evaluate(X_test, y_test)
```

## Classification Pipeline

For a smoother experimentation process, we created the ClassifierPipeline class, which is responsible for the entire workflow: building the dataset, training the model, evaluating the model, and visualizing the results.

```python
class ClassifierPipeline:
    """ClassifierPipeline defines the comon pipeline for all classification tasks:
        1 - build the dataset parsing folders and extracting features
        2 - training the model
        3 - evaluating the model
        4 - visualising the result
    """
    def __init__(self):
        """
        Initializes the ClassifierPipeline with dataset builder, feature extractor, 
        and classifier.
        """
        self.dataset_builder = None
        self.extractor = None
        self.classifier = None

    def run(self):
        """ Executes the classification pipeline: """
        
        # Step 1: Build Dataset
        # If the dataset is already built and ready, use it; otherwise, build the dataset from scratch.
        print("Building dataset...")
        if self.dataset_builder.ready:
            X_train, y_train, X_test, y_test = self.dataset_builder.get_dataset()
        else:
            X_train, y_train, X_test, y_test = self.dataset_builder.build_dataset(self.extractor)

        # Step 2: Train Model
        print("Training model...")
        # (num_mfccs, num_observations)
        input_shape = X_train.shape[1:] 
        
        # Get number of classes
        num_classes = len(np.unique(y_train))

        # Initialize and train the model
        classifier.init_model(input_shape, num_classes)
        classifier.train(X_train, y_train)
    
        # Step 3: Evaluate Model
        print("Evaluating model...")
        classifier.evaluate(X_test, y_test)

        # Step 4: Visualise result
        classifier.visualise()
        
```

Furthermore, ML classifiers tend to perform well when the features are independent, which is to some extent true for MFCCs, contributing to the good performance of both models.
